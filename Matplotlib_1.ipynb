# syntax = plt.plot(x,y) :- create a line plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8]
y=[21,25,19,27,24,31,25,29]

plt.plot(x,y)
```

```python
# syntax - plt.bar(labels,values)

animals = ['cat', 'dog', 'goat', 'sheep', 'horse', 'rabbit']
animal_count = [23, 30, 40, 38, 15, 21]
plt.bar(animals, animal_count, color='green')
```

```python
experience = [1,2,3,4,5]
salaries = [23000,21000,28000,32000,33000]
plt.plot(experience, salaries, color='#3B1E54', marker='o')
plt.xlabel('Experience (in years)')
plt.ylabel('Per Month Salary (in Rs.)')
plt.title('Experience vs. Salaries')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(2,2))

animals = ['cat', 'dog', 'goat', 'sheep', 'horse', 'rabbit']
animal_count = [23, 30, 40, 38, 15, 21]
plt.bar(animals, animal_count)

plt.xticks(rotation=60)
plt.yticks(rotation=90)

plt.savefig('Animal Count Plot')
plt.show()
```

## Multiple plots on a single chart

```python
age = [20,21,22,23,24,25]
weight_groupA = [38,39,41,37,43,38]
weight_groupB = [47,43,45,41,47,46]

plt.plot(age, weight_groupA, color='pink', marker='o', label='GroupA')
plt.plot(age, weight_groupB, color='blue', marker='*', label='GroupB')

plt.xlabel('Age (in years)')
plt.ylabel('Weight (in Kg)')
plt.title('Plot of Age vs. Weight for Two Groups')

plt.legend()
plt.show()
```

## Setting label custom location

```python
age = [20,21,22,23,24,25]
weight_groupA = [38,39,41,37,43,38]
weight_groupB = [47,43,45,41,47,46]

plt.plot(age, weight_groupA, color='yellow', marker='o', label='GroupA')
plt.plot(age, weight_groupB, color='brown', marker='x', label='GroupB')

plt.xlabel('Age (in years)')
plt.ylabel('Weight (in Kg)')
plt.title('Plot of Age vs. Weight for Two Groups')

# plt.legend(loc='upper right')
plt.legend(loc=(-0.5,0.5))    # numeric positions can also be used 
plt.show()
```

```python
# Ques. Display the top 5 games based on the highest global sales.
df_sorted = df_game.sort_values('Global_Sales', ascending=False)
```

```python
df_sorted
```

```python
games = df_sorted.iloc[:5, 1].values
games
```

```python
global_sales = df_sorted.iloc[:5, -1].values
global_sales
```

```python
plt.plot(games, global_sales, color='blue')
plt.xticks(rotation=60)
plt.xlabel('Game')
plt.ylabel('Global Sales')
plt.show()
```

or this can be modified to visualize better

```python
plt.figure(figsize=(10,8))
plt.plot(games, global_sales, color='green', marker='o')
plt.title('Top 5 Games')
plt.xlabel('Game')
plt.ylabel('Global Sales')
plt.grid()
plt.show()
```

```python
# Ques. Display the count of each genre
df_game['Genre'].value_counts()
```

```python
labels = df_game['Genre'].value_counts().index   #doubt
labels
```

```python
values = df_game['Genre'].value_counts().values
values
```

```python
plt.bar(labels, values)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.bar(labels, values, color='#B59F78')
plt.xticks(rotation=75, fontsize=12)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Video Game Count Genre wise')
plt.show()
```

```python
# to increase or decrease the size of bar width

plt.bar(labels, values, width=0.5, color='#2A3663')
plt.xticks(rotation=75)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Video Game Count Genre wise')
plt.show()
```

## seaborn

same kind of plot can be created using seaborn

```python
import seaborn as sns
```

```python
sns.barplot(x=labels , y=values , color = 'red' , width = 0.5)
plt.xticks(rotation=45)
plt.title("Video Game count Genre Wise")
plt.show()
```

```python
# Ques. Which platforms have the most games released over the years?

df_game['Platform'].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Platform',data=df_game,color='lightgreen')
plt.xticks(rotation=75)
plt.title('Number of Games Released By Platform')
plt.show()
```

The above graph can be visualized in order of the number of games released (in decreasing order) using the order parameter

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Platform', data=df_game, order=df_game['Platform'].value_counts().index, color='skyblue')
plt.xticks(rotation=75)
plt.title('Number of Games Released by Platform')
plt.xlabel('Number of Games')
plt.ylabel('platform')
plt.show()
```

or the graph can be rotated for better visualization

```python
plt.figure(figsize=(10,10))
sns.countplot(y='Platform', data=df_game, order=df_game['Platform'].value_counts().index, color='Red')
plt.title('Number of Games Released by Platform')
plt.xlabel('Number of Games')
plt.ylabel('platform')
plt.show()
```
